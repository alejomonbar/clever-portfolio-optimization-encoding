# 2. Portfolio optimization for S&P500 stocks

In this section, it's presented the solution of portfolio optimization using real data of the [SP500](https://en.wikipedia.org/wiki/S%26P_500) stock market index. We use the python based library [yfinance](https://pypi.org/project/yfinance/) to get updated stock prices during the last three months. The solution is presented for the portfolio optimization with budget inequality constraint $\sum_{i=1}^n x_i \le C$ for the unbalanced penalization and slack encodings using QAOA and VQE. The stocks are taking randomly for a small dataset with 36 different stocks.

## 1. Model development

Based on the portfolio optimization equations with the inequality constraint function, I use three functions 'portfolio_optimization' which creates a docplex model of the portfolio optimization problem, 'getting_data_stocks' to extract the S&P 500 stock prices and postprocess the expected return $\mu$ and the covariance matrix $\sigma$, and 'solution_str' with takes the docplex optimization model and solve it classically.

```python
def portfolio_optimization(mu, sigma, risk, budget=None, min_profit=None):
    
    mdl = Model('Portfolio Optimization')
    N_stocks, N_days = mu.shape
    # Binary set of variables that represent the stocks
    x = np.array(mdl.binary_var_list(N_stocks, name="stock")) # x vector in numpy array for matrix multiplication
    
    # Portfolio optimization function
    objective_function = (mu.mean(axis=1) @ x - risk * x.T @ sigma @ x) / np.mean(np.abs(mu.mean(axis=1)))
    mdl.maximize(objective_function)
    
    # Budget constraint
    if budget:
        mdl.add_constraint(mdl.sum(x) <= budget, ctname='budget')
    return mdl

def getting_data_stocks(stocks, period="12mo"):
    tickers = yf.Tickers(stocks)

    # Price of all the stocks of this example
    Pt = np.array([tickers.tickers[stock].history(period=period)["Close"] for stock in stocks])
    mu = (Pt[:,1:] - Pt[:,:-1]) / Pt[:,:-1]
    
    N_stocks, N_days = mu.shape
    sigma = np.zeros((N_stocks, N_stocks))
    for i in range(N_stocks-1):
        for j in range(i, N_stocks):
            sigma[i,j] = ((mu[i] - mu[i].mean()) * (mu[j] - mu[j].mean())).sum()/(N_days - 1)
    sigma += sigma.T
    data_stocks = {}
    data_stocks["stocks"] = stocks
    data_stocks["Pt"] = Pt
    data_stocks["mu"] = mu
    data_stocks["num_stocks"] = N_stocks
    data_stocks["num_days"] = N_days
    data_stocks["sigma"] = sigma
    return data_stocks

def solution_str(mdl):
    docplex_sol = mdl.solve() # Solving the docplex model using cplex
    solution = ""
    for ii in mdl.iter_binary_vars():
        solution += str(int(np.round(docplex_sol.get_value(ii), 1)))
    return solution
```

```python
seed = 1
all_stocks = ["AAPL", "MSFT", "AMZN","TSL","GOOGL","GOOG", "NVDA","META","UNH","XOM","V", "MA", "CVX","PEP",
             "PFE","COST","WMT", "LIN", "CMCSA", "VZ", "TXN", "NKE", "NFLX", "ORCL", "COP", "HON", "LOW",
             "DE", "IBM", "ELV", "LMT", "AXP", "GILD", "C", "AMAT", "ADP", "MMC", "ZTS"]
cases = np.arange(5, 31)
np.random.seed(seed)
random.seed(seed)
info_stocks = {}
for num_stocks in tqdm(cases):
    stocks = random.sample(all_stocks, num_stocks)
    info_stocks[num_stocks] = getting_data_stocks(stocks, period="3mo")
    budget = np.floor(0.7 * num_stocks) 
    mu = info_stocks[num_stocks]["mu"]
    sigma = info_stocks[num_stocks]["sigma"]
    info_stocks[num_stocks]["mdls"] = portfolio_optimization(mu, sigma, risk=0.1, budget=budget)
    info_stocks[num_stocks]["sol_str"] = solution_str(info_stocks[num_stocks]["mdls"])
    info_stocks[num_stocks]["budget"] = budget

np.save("./Data/setup_real_stocks.npy", info_stocks)
```

```python
from openqaoa.workflows.optimizer import QAOA
from openqaoa.problems.converters import FromDocplex2IsingModel # function to translate docplex problems into Ising Hamiltonians

def qaoa_results(penalty, mdl, sol_str, p=1, t=0.1, maxiter=100, unbalanced=False):
    """
    Solving a docplex model using QAOA in OpenQAOA
    
    penalty: The penalty terms for the equality and inequality constraints.
    
    
    
    """
    ising_hamiltonian = FromDocplex2IsingModel(mdl, multipliers=penalty[0],
                                   unbalanced_const=unbalanced,
                                   strength_ineq=[penalty[1],penalty[2]]).ising_model

    qaoa = QAOA() # OpenQAOA QAOA model
    qaoa.set_circuit_properties(p=p, init_type="ramp", linear_ramp_time=t) # initialization betas and gammas with a ramp technique
    qaoa.set_classical_optimizer(maxiter=maxiter) 
    qaoa.compile(ising_hamiltonian)
    qaoa.optimize()
    nq = ising_hamiltonian.n
    results = qaoa.results.lowest_cost_bitstrings(2**nq)
    results["openqaoa"] = qaoa.results
    for nn, string in enumerate(results["solutions_bitstrings"]):
        if string[:len(sol_str)] == sol_str:
            pos = nn
            break
    results["opt_pos"] = pos # Optimal position in the sorted Hamiltonian
    results["probability"] = results["probabilities"][results["opt_pos"]] # Probability of the optimal solution
    results["classical_sol"] = sol_str # classical solution
    results["CoP"] = results["probability"] * 2 ** nq # Coefficient of Performance https://arxiv.org/abs/2211.13914
    results["Emin"] = np.min(results["bitstrings_energies"]) # minimum eigenvalue
    results["Emax"] = np.max(results["bitstrings_energies"]) # maximum eigenvalue
    results["Ef"] = np.min(qaoa.results.optimized["optimized cost"]) # current average energy
    results["r"] = (results["Ef"] - results["Emax"])/(results["Emin"] - results["Emax"]) # Approximation ratio
    results["opt_angles"] = qaoa.results.optimized["optimized angles"] # Optimal betas and gammas 
    results["num_qubits"] = nq
    results["ising_hamiltonian"] = ising_hamiltonian
    results["cx"] = 2 * len([t for t in ising_hamiltonian.terms if len(t) == 2]) * p
    return results

def find_penalization_terms(penalty, mdls, list_assets, sol_str):
    cop = 0 # Coefficient of performance (one of our objectives) See at the end of Sec.III for an explanation
    opt_positions = 0
    for asset in list_assets:
        results = qaoa_results(penalty, mdls[asset], sol_str[asset], p=1, t=0.1, unbalanced=True, maxiter=100)
        opt_pos = results["opt_pos"] # The position of the optimal solution on the sorted Ising Hamiltonian of the problem (ideally pos=0)
        cop += results["CoP"]  # Coefficient of performance (It quantifies how many times it is better to find the optimal solution compared with a random guessing)
        opt_positions += opt_pos
        print(f"Num. assets: {asset} | penalty:{penalty} | CoP:{results['CoP']} | optimal position {results['opt_pos']}")
    return -cop + 10*opt_positions
```

```python
info_stocks = np.load("./Data/setup_real_stocks.npy", allow_pickle=True).item()
mdls = {i:info_stocks[i]["mdls"] for i in range(5,31)}
sol_str = {i:info_stocks[i]["sol_str"] for i in range(5,31)}

penalty0 = [0.5,0.1,0.3] # initial guess
solf = minimize(find_penalization_terms,
                x0=penalty0,
                args=(mdls, [6,8], sol_str),
                bounds=((1e-2,10),(1e-3,2),(1e-3,2)),
                method="powell",
                options={"maxiter":100}) #Minimizing the objective function that depends on the encoding of the optimal solution and the probability of getting that
```

```python
from openqaoa.devices import create_device

def qaoa_results(penalty, mdl, sol_str, p=1, t=0.1, shots=100, maxiter=100, unbalanced=False,
                 device='arn:aws:braket:::device/quantum-simulator/amazon/sv1'):
    """
    Solving a docplex model using QAOA in OpenQAOA
    
    penalty: The penalty terms for the equality and inequality constraints.
    
    """
    ising_hamiltonian = FromDocplex2IsingModel(mdl, multipliers=penalty[0],
                                   unbalanced_const=unbalanced,
                                   strength_ineq=[penalty[1],penalty[2]]).ising_model
    braket_device = create_device(location='aws', name=device_name)
    qaoa = QAOA() # OpenQAOA QAOA model
    qaoa.set_device(braket_device)
    qaoa.set_backend_properties(n_shots=shots)
    qaoa.set_circuit_properties(p=p, init_type="ramp", linear_ramp_time=t) # initialization betas and gammas with a ramp technique
    qaoa.set_classical_optimizer(maxiter=maxiter) 
    qaoa.compile(ising_hamiltonian)
    qaoa.optimize()
    nq = ising_hamiltonian.n
    results = qaoa.results.lowest_cost_bitstrings(n_shots)
    results["openqaoa"] = qaoa.results
    for nn, string in enumerate(results["solutions_bitstrings"]):
        if string[:len(sol_str)] == sol_str:
            pos = nn
            break
        else:
            pos = -1
    results["opt_pos"] = pos # Optimal position in the sorted Hamiltonian
    results["probability"] = results["probabilities"][results["opt_pos"]] # Probability of the optimal solution
    results["classical_sol"] = sol_str # classical solution
    results["CoP"] = results["probability"] * 2 ** nq # Coefficient of Performance https://arxiv.org/abs/2211.13914
    results["Emin"] = np.min(results["bitstrings_energies"]) # minimum eigenvalue
    results["Emax"] = np.max(results["bitstrings_energies"]) # maximum eigenvalue
    results["Ef"] = np.min(qaoa.results.optimized["optimized cost"]) # current average energy
    results["r"] = (results["Ef"] - results["Emax"])/(results["Emin"] - results["Emax"]) # Approximation ratio
    results["opt_angles"] = qaoa.results.optimized["optimized angles"] # Optimal betas and gammas 
    results["num_qubits"] = nq
    results["ising_hamiltonian"] = ising_hamiltonian
    results["cx"] = 2 * len([t for t in ising_hamiltonian.terms if len(t) == 2]) * p
    return results
```

### Landscape analysis

In this section, I'm going to test the IonQ device comparing its landscape from the landscape of the simulator. The landscape in this case correspong to the energy expectation value from a range of gamma $\{-\pi, 0\}$ and beta $\{-\pi/2, 0\}$. In this case, I took the 5 stocks case.

```python
betas = np.linspace(-np.pi/2, 0 , 20)
gammas = np.linspace(-np.pi, 0, 20)
energy = np.zeros((len(betas), len(gammas)))
for i, gamma in enumerate(gammas):
    for j, beta in enumerate(betas):
        params = [beta, gamma]
        varational_params = q.variate_params
        varational_params.update_from_raw(params)
        energy[i,j] = q.backend.expectation(varational_params)
        print(f"i:{i} | j:{j} | Energy:{energy[i,j]}")
np.save("./Data/energy_landscape_sv1.npy", energy)
```
