# Portfolio optimization: Comparing unbalanced penalization vs. Quantum Zeno Dynamics

Portfolio optimization problems usually come with a set of constraints, for example, in [[1]](https://arxiv.org/pdf/2209.15024.pdf) two inequality constraints are applied to ensure the investment in the assets is below the budget and to guarantee a minimum profit (see Sec. IV A of [[1]](https://arxiv.org/pdf/2209.15024.pdf)). Usually, to encode these inequality constraints in quantum computers we require extra qubits (for the slack variables). This has two disadvantages, first the probability of finding optimal solutions decrease and second the number of qubits and the depth of the circuits increases. To solve this problem, I use unbalanced penalization [[2]](https://arxiv.org/pdf/2211.13914.pdf), a new encoding method that does not require those extra qubits for the inequality constraints. This technique has been used in the traveling salesman problem, the bin packing problem, and the knapsack problem. I extended the solution to work with portfolio optimization. This method is expected to enhance the probability of finding optimal solutions, reducing the depth of the circuit and the number of qubits required. I solve the problem using QAOA, and VQE and compare it with the results from [1]. Additionally, results on real hardware are presented, and a visualization technique for the optimal solution and the circuit requirements.

In this section, it is used the unbalanced penalization method [[2]](https://arxiv.org/pdf/2211.13914.pdf) to encode different portfolio optimization problems subject to the following conditions 

* Cost function

$$\large{\min_{x \in \{0,1\}^{n}} q  x^T\Sigma x - \mu^T x}.$$

* Budget constraint:

$$\sum_j^n x_j \le C.$$

* Minimum return constraint

$$\sum_j^n \mu_j x_j \ge R.$$

where $x$ is a vector with elements $x_i=1$ if stock i is choosen and 0 otherwise, $n$ is the number of assets, $q$ is the risk factor, $\Sigma$ is covariance matrix, $\mu$ is the expected return, $C$ is the budget, and $R$ is the minimum return. 

The unbalanced penalization method takes inequality constraints of the form 

$$h(\mathrm{x}) =  \sum_i l_i x_i - B \ge 0$$

and translate them into

$$\zeta(x) = - \lambda_1 h(x) + \lambda_2 h(x)^2  = - \lambda_1\left(\sum_i l_i x_i - B\right) + \lambda_2 \left(\sum_i l_i x_i - B\right)^2 ,$$

where $\lambda_{1,2}$ are penalization terms that must be tuned for the specific problem (Details of the derivation of the unbalanced penalization method can be found in Sec. III of [[2]](https://arxiv.org/pdf/2211.13914.pdf)). For the portfolio optimization, the QUBO is given by

$$\min_{x \in \{0,1\}^{n}} q  x^T\Sigma x - \mu^T x - \lambda_1\left(C - \sum_j^n x_j\right) + \lambda_2\left(C - \sum_j^n x_j\right) - \lambda_1\left(\sum_j^n \mu_j x_j - R\right) + \lambda_2\left(\sum_j^n \mu_j x_j - R\right),$$

This solution gets rid of the slack variables usually used for inequality constraints of QUBOs. It has the advantage of reducing the number of qubits and gates needed and it shows a huge improvement in getting the optimal solution as it is shown in Sec. IV of [[2]](https://arxiv.org/pdf/2211.13914.pdf).

## 1. Model development

Based on the equations presented before, here there are two functions 'portfolio_optimization' which creates a docplex model of the portfolio optimization problem. This step is important because from docplex we can extract the optimal solution solving the problem with equation 'solution_str'. 

```python
def portfolio_optimization(mu, sigma, risk, budget=None, min_profit=None):
   
    mdl = Model('Portfolio Optimization')
    N_stocks = len(mu)
    # Binary set of variables that represent the stocks
    x = np.array(mdl.binary_var_list(N_stocks, name="stock")) # x vector in numpy array for matrix multiplication
    
    # Portfolio optimization function
    objective_function = (mu @ x - risk * x.T @ sigma @ x) / np.max(np.abs(mu))
    mdl.maximize(objective_function)
    
    # Budget constraint
    if budget:
        mdl.add_constraint(mdl.sum(x) <= budget, ctname='budget')
    # Minimum return 
    if min_profit:
        mdl.add_constraint(mdl.sum(mui * xi / min_profit for mui, xi in zip(mu, x)) >= 1)
    return mdl

def solution_str(mdl):
    docplex_sol = mdl.solve() # Solving the docplex model using cplex
    solution = ""
    for ii in mdl.iter_binary_vars():
        solution += str(int(np.round(docplex_sol.get_value(ii), 1)))
    return solution
```

## 2. Extracting the models from the paper: Portfolio Optimization via Quantum Zeno Dynamics on a Quantum Processor

```python
from problem_instances import get_instance
list_assets = np.arange(4, 11)
mdls_1 = {}
mdls_2 = {}

sol_str_1 = {}
sol_str_2 = {}

for num_assets in list_assets:# number of assets to choose from
    inst = get_instance("multi", num_assets)
    mu = inst["mu"]  # expected return for each asset in the pool
    sigma = inst["sigma"]  # covariance matrix of returns for the assets in the pool
    budget = inst["budget"]  # maximum number of assets allowed in the portfolio
    min_return = inst["min_return"]  # minimum total expected return of the portfolio
    risk = inst["risk_factor"]
    # Docplex models for the two different problems
    mdls_1[num_assets] = portfolio_optimization(mu, sigma, risk, budget) #Only budget constraint
    mdls_2[num_assets] = portfolio_optimization(mu, sigma, risk, budget, min_return) #Budget and minimum return constraints
    # Solution of the two different problems
    sol_str_1[num_assets] = solution_str(mdls_1[num_assets])
    sol_str_2[num_assets] = solution_str(mdls_2[num_assets])
```

### Tuning the penalization term

For tuning the penalty terms $\lambda_0$, $\lambda_1$, and $\lambda_2$ of the unbalanced penalization approach, I use an objective function that depends on two factors, first, the encoding of the optimal solution on the ground state of the Ising Hamiltonian that represents the portfolio optimization problem and second, the probability of getting the optimal solution (cplex solution) of the problem using QAOA with p=1 must be as large as possible. I use openqaoa because it has a predefined function to translate docplex models into Ising Hamiltonians, it has implemented the unbalanced penalization method, and the code is tailored to run quickly QAOA problems.

```python
from openqaoa.workflows.optimizer import QAOA
from openqaoa.problems.converters import FromDocplex2IsingModel # function to translate docplex problems into Ising Hamiltonians

def qaoa_results(penalty, mdl, sol_str, p=1, t=0.1, maxiter=100, unbalanced=False):
    """
    Solving a docplex model using QAOA in OpenQAOA
    
    penalty: The penalty terms for the equality and inequality constraints.
    
    
    
    """
    ising_hamiltonian = FromDocplex2IsingModel(mdl, multipliers=penalty[0],
                                   unbalanced_const=unbalanced,
                                   strength_ineq=[penalty[1],penalty[2]]).ising_model

    qaoa = QAOA() # OpenQAOA QAOA model
    qaoa.set_circuit_properties(p=p, init_type="ramp", linear_ramp_time=t) # initialization betas and gammas with a ramp technique
    qaoa.set_classical_optimizer(maxiter=maxiter) 
    qaoa.compile(ising_hamiltonian)
    qaoa.optimize()
    results = qaoa.results.lowest_cost_bitstrings(2**mdl.number_of_binary_variables)
    results["openqaoa"] = qaoa.results
    results["opt_pos"] = results["solutions_bitstrings"].index(sol_str) # Optimal position in the sorted Hamiltonian
    results["probability"] = results["probabilities"][results["opt_pos"]] # Probability of the optimal solution
    results["classical_sol"] = sol_str # classical solution
    results["CoP"] = results["probability"] * 2 ** len(sol_str) # Coefficient of Performance https://arxiv.org/abs/2211.13914
    results["Emin"] = np.min(results["bitstrings_energies"]) # minimum eigenvalue
    results["Emax"] = np.max(results["bitstrings_energies"]) # maximum eigenvalue
    results["Ef"] = np.min(qaoa.results.optimized["optimized cost"]) # current average energy
    results["r"] = (results["Ef"] - results["Emax"])/(results["Emin"] - results["Emax"]) # Approximation ratio
    results["opt_angles"] = qaoa.results.optimized["optimized angles"] # Optimal betas and gammas 
    return results

def find_penalization_terms(penalty, mdls, list_assets, sol_str):
    cop = 0 # Coefficient of performance (one of our objectives) See at the end of Sec.III for an explanation
    opt_positions = 0
    for asset in list_assets:
        results = qaoa_results(penalty, mdls[asset], sol_str[asset], p=1, t=0.1, unbalanced=True, maxiter=100)
        opt_pos = results["opt_pos"] # The position of the optimal solution on the sorted Ising Hamiltonian of the problem (ideally pos=0)
        cop += results["CoP"]  # Coefficient of performance (It quantifies how many times it is better to find the optimal solution compared with a random guessing)
        opt_positions += opt_pos
        print(f"Num. assets: {asset} | penalty:{penalty} | CoP:{results['CoP']} | optimal position {results['opt_pos']}")
    return -cop + 10*opt_positions
```

### Minimization function

I use the scipy minmization function to find the penalization terms using the method 'Powell' and three different cases, the portfolio optimization with 4, 6, and 8 assets. Note, that as it is shown also in the paper [2], the unblanaced penalization method shows good generalization capabilities, once found a set of penalization terms they works well for the other problems.

```python
penalty0 = [0.5,0.1,0.3] # initial guess
solf = minimize(find_penalization_terms,
                x0=penalty0,
                args=(mdls_2, [4,6,8], sol_str_2),
                bounds=((1e-2,10),(1e-3,2),(1e-3,2)),
                method="powell",
                options={"maxiter":100}) #Minimizing the objective function that depends on the encoding of the optimal solution and the probability of getting that
```

## CoP: Coefficient of performance
The coefficient of performance (CoP), a quantifier that expresses how good a quantum algorithm is to give a specific solution compared to random guessing over a set of solutions. In this perspective, we use the CoP to determine how likely it is to find the optimal solution for portfolio optimization on the local minima of the QAOA algorithm for differen $p = \{1,5\}$. The CoP is given by

$$CoP = \frac{P(^*\mathrm{x})}{P_R},$$

where $P(^*\mathrm{x})$ is the probability of getting a bit string $^*\mathrm{x}$ using a quantum algorithm, e.g, QAOA and $P_R =1 / 2^n$ is the probability of a random guessing with n equal to the number of qubits involved in the problem.


```python
colors = plt.cm.get_cmap("jet", 12)

prop = "CoP"
prop_dict = {"mdl_1":{}, "mdl_2":{}}
for n_assets in list_assets:
    prop_dict["mdl_1"][n_assets] = []
    prop_dict["mdl_2"][n_assets] = []
    for i in range(1,6):
        prop_dict["mdl_1"][n_assets].append(results_qaoa["mdl_1"][n_assets][i][prop])
        prop_dict["mdl_2"][n_assets].append(results_qaoa["mdl_2"][n_assets][i][prop])
        
fig, ax = plt.subplots(1,2, figsize=(14,5))
for n in list_assets:
    ax[0].plot(range(1,6), prop_dict["mdl_1"][n], label=f"{n} assets ", marker="o", markeredgecolor="black", color=colors(n))
    ax[1].plot(range(1,6), prop_dict["mdl_2"][n], label=f"{n} assets ", marker="o", markeredgecolor="black", color=colors(n))
for i in range(2):
    ax[i].set_xlabel("p")
    ax[i].set_xticks(range(1,6))
    ax[i].grid()
ax[0].set_ylabel(f"{prop}")
ax[0].legend(loc='upper center', bbox_to_anchor=(1.1,1.2), ncol=8)
fig.savefig(f"./Images/unbalanced_penalization_assets_{prop}.png",dpi=500, transparent=True)
```

## Sorting the energy eigenvalues

I sorted the energy eigenvalues of the cost Hamiltonian for the different portfolio optimization problems. The parameter of interest is the position of the energy eigenvalue that describes the optimal solution of the original  portfolio optimization problem within the list of all sorted energy eigenvalues of the cost Hamiltonian of their QUBO representation. If the position of the optimal solution is far from the ground energy of the cost Hamiltonian, the unbalanced penalization encoding is not working as expected. Otherwise, if the energy eigenvalue corresponding to the optimal solution is close to the ground state energy given the total number of eigenvalues, the unbalanced encoding is advantageous. In the portfolio optimization cases I used here, the worst case for encoding the optimal solution was for 9 assets with the optimal solution encoded in the 4th out of 512 eigenvalues.

```python
colors = plt.cm.get_cmap("jet", 12)

prop = "opt_pos"
prop_dict = {"mdl_1":{}, "mdl_2":{}}
for n_assets in list_assets:
    prop_dict["mdl_1"][n_assets] = []
    prop_dict["mdl_2"][n_assets] = []
    for i in range(1,6):
        prop_dict["mdl_1"][n_assets].append(results_qaoa["mdl_1"][n_assets][i][prop])
        prop_dict["mdl_2"][n_assets].append(results_qaoa["mdl_2"][n_assets][i][prop])
        
fig, ax = plt.subplots(1,2, figsize=(14,5))
for n in list_assets:
    ax[0].plot(n, prop_dict["mdl_1"][n][0], marker="o", markeredgecolor="black", color=colors(n), markersize=15)
    ax[1].plot(n, prop_dict["mdl_2"][n][0], marker="o", markeredgecolor="black", color=colors(n), markersize=15)
for i in range(2):
    ax[i].set_xlabel("Num. assets")
    ax[i].set_xticks(list_assets)
    ax[i].grid()
ax[0].set_ylabel("Optimal solution eigenvalue pos")
fig.savefig(f"./Images/unbalanced_penalization_assets_{prop}.png",dpi=500, transparent=True)
```
