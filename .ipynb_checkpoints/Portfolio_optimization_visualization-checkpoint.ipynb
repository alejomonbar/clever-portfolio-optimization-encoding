# 3. Visualization of the portfolio optimization

One of the main problems in general with combinatorial optmization problems after been solved by quantum computing algorithms is the visualization stage. For example in the portfolio optimization, after getting a string with zeros and ones, there is one step left and it is know if the solution looks right how it compares with other solutions and so on. 

In this section, I try to answer this question with a visualization tool called 'portfolio_optimization_visualization', a graph representation of portfolio optimmization. Fig. 1 shows the results of the visualization of one problem with 17 stocks from the S&P 500.
```python
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx

label_size = 12
plt.rcParams['xtick.labelsize'] = label_size 
plt.rcParams['ytick.labelsize'] = label_size 
plt.rcParams['axes.labelsize'] = label_size 
plt.rcParams['legend.fontsize'] = label_size 

%matplotlib inline
# From Sec. 2 Portfolio optimization for S&P500 stocks data
setup = np.load("./Data/setup_real_stocks.npy", allow_pickle=True).item() 
qaoa_results = np.load("./Data/results_qaoa_local.npy", allow_pickle=True).item() 
```

```python
def portfolio_optimization_visualization(stocks, setup, qaoa_results):
    colors = plt.cm.get_cmap("jet", 12)
    cost_hamiltonian = qaoa_results["unbalanced"][stocks][1]["openqaoa"].cost_hamiltonian
    mu = setup[stocks]["mu"]
    sigma = setup[stocks]["sigma"]
    sigma_norm = 20*sigma/ np.abs(np.max(sigma))
    Pt = setup[stocks]["Pt"]
    fig, ax = plt.subplots(2,2,figsize=(12,11))

    # -------------- Stocks behaviour
    for i in range(stocks):
        ax[0,0].plot(Pt[i], color=colors(i), label=setup[stocks]["stocks"][i])
    ax[0,0].set_xlabel("day")
    ax[0,0].set_ylabel("price")
    ax[0,0].legend(loc='upper center', bbox_to_anchor=(0.5,1.3), ncol=6)
    # ----------- Setup of the problem 
    G = nx.Graph()
    G.add_nodes_from(range(stocks))
    edge_list = []
    width = []
    for i in range(stocks-1):
        for j in range(i+1, stocks):
            G.add_edge(i,j, weight=sigma[i,j])
            edge_list.append((i,j))
            width.append(sigma_norm[i,j])
    pos = nx.circular_layout(G)
    graph_colors = [colors(8) if i else colors(3) for i in mu.mean(axis=1)>0]
    nx.draw(G, pos=pos, labels = {i:n for i, n in enumerate(setup[stocks]["stocks"])},
            node_size=1200*np.abs(mu.mean(axis=1))/np.max(np.abs(mu.mean(axis=1))),node_color=graph_colors,ax=ax[0,1],
           edgecolors="grey")

    nx.draw_networkx_edges(G, pos, edgelist=edge_list, width=width, ax=ax[0,1])
    ax[0,1].plot([],[], marker="o", linewidth=0, markeredgecolor="grey",color=colors(3), label="$\mu$ -", markersize=10)
    ax[0,1].plot([],[], linewidth=1, color="black", label="$\downarrow\sigma$", markersize=10)
    ax[0,1].plot([],[], marker="o", linewidth=0, markeredgecolor="grey",color=colors(8), label="$\mu$ +", markersize=10)
    ax[0,1].plot([],[], linewidth=3.5, color='black', label="$â†‘\sigma$", markersize=10)

    ax[0,1].legend(loc='upper center', bbox_to_anchor=(0.5,1.25), ncol=2)
    # ----------- Solution Graph
    graph_colors_sol = [colors(9) if int(i) else colors(4) for i in setup[stocks]["sol_str"]]

    nx.draw(G, pos=pos, labels = {i:n for i, n in enumerate(setup[stocks]["stocks"])},
            node_size=900,node_color=graph_colors_sol,ax=ax[1,0],
           edgecolors="grey")
    nx.draw_networkx_edges(G, pos, edgelist=edge_list, width=width, ax=ax[1,0])
    ax[1,0].plot([],[], marker="o", linewidth=0, markeredgecolor="grey",color=colors(4), label="selected", markersize=10)
    ax[1,0].plot([],[], marker="o", linewidth=0, markeredgecolor="grey",color=colors(9), label="no selected", markersize=10)

    ax[1,0].legend(loc='upper center', bbox_to_anchor=(0.5,1.08), ncol=2)

    # ----- Frontier
    mu_T = mu.mean(axis=1)
    for sol in qaoa_results["unbalanced"][stocks][5]["solutions_bitstrings"][:1000]:
        x = np.array([[int(i) for i in sol]]).T
        profit = mu_T @ x
        risk = x.T @ sigma @ x
        if sol == setup[stocks]["sol_str"]:
            color = colors(9)
            markersize=10
        else:
            color = "grey"
            markersize=5
        ax[1,1].plot(risk[0], [profit], marker="o", color=color, markersize=markersize, markeredgecolor="black")
    ax[1,1].plot([], [], marker="o",linewidth=0, color=colors(9), markersize=10, markeredgecolor="black", label="optimal")
    ax[1,1].plot([], [], marker="o",linewidth=0, color="grey", markersize=5, markeredgecolor="black", label="other")
    ax[1,1].set_xlabel("risk")
    ax[1,1].set_ylabel("profit")
    ax[1,1].legend()
    fig.savefig(f"./Images/visualization_{stocks}.png", dpi=500, transparent=True, bbox_inches="tight")
```
